```python
data=data.rename(columns={"Annual Income (k$)":"Annual_Income","Spending Score (1-100)":"Spending_Score"})
```

```python
#Annual Income (k$)
#Spending Score (1-100)
```

```python
experimento=data[["Age","Annual_Income","Spending_Score"]]
experimento.head()
```

```python
# 
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(experimento)
print('type:',type(clusters))
clusters
```

```python
clusters=pd.DataFrame(clusters)
clusters=clusters.rename(columns={0:"cluster"})
```

```python
clusters
```

```python
clusters["cluster"].value_counts()
```

```python
data_total=pd.concat([data,clusters],axis=1)
data_total.head()
```

```python
#data_total=data_total["cluster"]+1
data_total["cluster"] = data_total["cluster"] + 1
data_total.shape
```
